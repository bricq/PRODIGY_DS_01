```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Internship\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_637467.csv")
```

```python
print(data.head())
```

```python
# Bar chart for the categorical variable 'Region'
region_counts = data['Region'].value_counts()

plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Region Distribution in Population')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart for the categorical variable 'IncomeGroup'
income_group_counts = data['IncomeGroup'].value_counts()

plt.figure(figsize=(10, 6))
income_group_counts.plot(kind='bar', color='lightgreen')
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.title('Income Group Distribution in Population')
plt.xticks(rotation=45)
plt.show()
```
